```python
from matplotlib import pyplot as plt
import numpy as np
```

ZADANIE 1

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')

plt.title('Wykresy sinus i cosinus')
plt.xlabel('x')
plt.ylabel('y')

plt.xlim(-2 * np.pi, 2 * np.pi)
plt.ylim(-1.25, 1.25)

plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.legend()

plt.show()
```

ZADANIE 3

```python
import random
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(0,1)
ax.set_ylim(0,1)
n = random.randint(3,10)

for _ in range(n):
    figura = random.choice(['kolo', 'kwadrat', 'trojkat'])
    x = random.random()
    y = random.random()
    kolor = np.random.rand(3,)

    if figura == 'kolo':
        kolo = plt.Circle((x, y), 0.1, color=kolor)
        ax.add_patch(kolo)
    elif figura == 'kwadrat':
        kwadrat = plt.Rectangle((x, y), 0.3, 0.3, color=kolor)
        ax.add_patch(kwadrat)
    else:

        rozmiar = 0.15
        wierzcholki = np.array([
            [x, y + rozmiar],
            [x - rozmiar * np.sqrt(3)/2, y - rozmiar/2],
            [x + rozmiar * np.sqrt(3)/2, y + rozmiar/2]
        ])
        trojkat = plt.Polygon(wierzcholki, color=kolor)
        ax.add_patch(trojkat)

plt.show()


```

ZADANIE 4

```python
x = np.linspace(-20, 20, 500)

funkcje = [
    (lambda x: x, "f(x) = x"),
    (lambda x: x**2, "f(x) = x^2"),
    (lambda x: x**3, "f(x) = x^3"),
    (lambda x: x**4, "f(x) = x^4"),
]

fig, ax = plt.subplots(2, 2, figsize=(12, 8))

for ax, (f, opis) in zip(ax.flat, funkcje):
    y = f(x)
    ax.plot(x, y)
    ax.set_title(opis)
    ax.grid(True)
    ax.axhline(0, color='black')
    ax.axvline(0, color='black')
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
plt.show()
```

ZADANIE 5

```python
plec = {'kobieta':100, 'mczyzna':95}
wiek = {'<20':20, '21-30':40, '31-40':55, '41-50':121, '>51':61 }
wyksztalcenie = {'podstawowe':10, 'rednie':85, 'wysze': 123}

fig, ax = plt.subplots(3, 1, figsize=(12, 8))
ax[0].pie(
    plec.values(),
    labels=plec.keys(),
    autopct='%1.1f%%'
    )
ax[0].set_title('Pe')

ax[1].pie(
    wiek.values(),
    labels=wiek.keys(),
    autopct='%1.1f%%'
)
ax[1].set_title("Wiek")

ax[2].pie(
    wyksztalcenie.values(),
    labels = wyksztalcenie.keys(),
    autopct='%1.1f%%'
)
ax[2].set_title("Wyksztalcenie")

plt.show()
```
