```python
import math

def zad4(a):
    if a <= 1:
        return False
    else:
        for i in range(2, int(math.sqrt(a)) + 1):
            if a % i == 0:
                return False
        return True
print (zad4(15))
```

```python
def zad5(a):
    wynik = []
    for i in range(1, a + 1):
        if a % i == 0:
            wynik.append(i)
    return wynik
print(zad5(120))
```

```python
def zad6(a):
    f = 2
    wynik = []
    while f*f <= a:
        while a % f == 0:
            wynik.append(f)
            a //= f
        f+=1
    if a > 1:
        wynik.append(a)
    return wynik
print(zad6(120))
```

```python
def nwd(a, b):
    while b > 0:
        temp = a % b
        a = b
        b = temp % b
    return a
def zad7(a, b):
    return a*b/nwd(a, b)

print (zad7(4,6))
```

```python
def dzielniki_wlasc(a):
    wynik = []
    for i in range(1, a):
        if a % i == 0:
            wynik.append(i)
    return wynik

def suma_dzielnikow_wlasc(a):
    return sum(dzielniki_wlasc(a))

def zad8(a, b):
    suma1 = suma_dzielnikow_wlasc(a)
    suma2 = suma_dzielnikow_wlasc(b)
    if suma1 == b and suma2 == a :
        return True
    else:
        return False
print(zad8(220,284))
```

```python
def zad9(a):
    wynik = []
    for i in range(1, a):
        if suma_dzielnikow_wlasc(i) == i:
            wynik.append(i)
    return wynik

print(zad9(500))
```

```python
def zad10(a):
    wynik = ""
    while a:
        temp = a % 2
        wynik+=str(temp)
        a //= 2
    return wynik[::-1]
print(zad10(192))
```

```python
def zad11(a,b,c):
    delta = b*b - 4*a*c
    if a!= 0:
        if delta < 0:
            return False
        elif delta == 0:
            x1 = (-b / (2*a))
            return x1
        else:
            x1 = (-b + math.sqrt(delta)) / (2*a)
            x2 = (-b - math.sqrt(delta)) / (2*a)
            return x1, x2
    else:
        return False
print(zad11(-2,4,-2))
```

```python
def zad12_it(a):
    wynik = 1
    if a < 0:
        return False
    for i in range(1, a + 1):
        wynik *= i
    return wynik
print(zad12_it(5))

def zad12_rek(a):
    if a == 0:
        return 1
    elif a < 0:
        return False
    else:
        return a * zad12_rek(a-1)
print(zad12_rek(5))
```

```python
def zad15():
    tab = input().split()
    n = len(tab)
    suma = 0
    for i in range(n):
        suma += int(tab[i])
    return suma/n
zad15()
```

```python
def zad17(a):
    male = 0
    duze = 0
    for znak in a:
        if znak.islower():
            male += 1
        elif znak.isupper():
            duze += 1
    return male, duze
print (zad17("ABCDEFG abcdefg 1"))
```

```python
def zad18(a):
    unikalne = []
    for znak in a:
        if znak not in unikalne:
            unikalne.append(znak)
    return unikalne
print(zad18("ala ma kota 1"))
```
