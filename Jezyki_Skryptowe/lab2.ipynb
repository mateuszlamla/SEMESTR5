```python
class Konto:
    def __init__(self, wlasciciel: str, nr_konta: str, saldo: float, kredyt: float, lokata: float):
        self.wlasciciel = wlasciciel
        self.nr_konta = nr_konta
        self.saldo = float(saldo)
        self.kredyt = kredyt
        self.lokata = lokata

    def zaloz_lokata(self, kwota):
        if kwota > self.saldo or kwota <= 0:
            print("Kwota mniejsza niz 0 lub kwota wieksza niz saldo")
        else:
            self.saldo -= kwota
            self.lokata += kwota
            print(f"zalozono lokate z kwota {kwota}")

    def wez_kredyt(self, kwota):
        if kwota > 0:
            self.kredyt += kwota
            self.saldo += kwota
            print(f"Wzieto kredyt na kwote {kwota}")
        else:
            print("Kwota mniejsza niz 0")

    def przelew(self, kwota, konto_docel):
        if kwota > self.saldo or kwota <= 0:
            print("Kwota mniejsza niz 0 lub kwota wieksza niz saldo")
        else:
            self.saldo -= kwota
            konto_docel.saldo += kwota
            print(f"Wykonano przelew na konto {konto_docel.nr_konta} na kwote {kwota}")

    def __str__(self):
        return f"Wlasciciel: {self.wlasciciel}, Nr Konta: {self.nr_konta}, Saldo: {self.saldo}, Kredyt: {self.kredyt}, Lokata: {self.lokata}"

konto_1 = Konto("Jan Kowalski", "123456", 10000, 0, 0)
konto_2 = Konto("Pawel Nowak", "098765", 3000, 0, 0)

konto_1.zaloz_lokata(100)
konto_2.przelew(1000, konto_1)
print(konto_1)
print(konto_2)
print()
konto_1.wez_kredyt(500)
print(konto_1)
```

```python
class Konto_VIP(Konto):
    def __init__(self, wlasciciel: str, nr_konta: str, saldo: float, kredyt: float, lokata: float):
        super().__init__(wlasciciel, nr_konta, saldo, kredyt, lokata)
        self.historia = []

    def zaloz_lokata(self, kwota):
        super().zaloz_lokata(kwota)
        self.historia.append(f"Zalozono lokate {kwota} zł")

    def wez_kredyt(self, kwota):
        super().wez_kredyt(kwota)
        self.historia.append(f"Wzięto kredyt {kwota} zł")

    def przelew(self, kwota, konto_docel):
        super().przelew(kwota, konto_docel)
        self.historia.append(f"Wykonano przelew na konto {konto_docel.nr_konta}, Kwota: {kwota} zł")

    def historia_konta(self):
        return self.historia


konto_3 = Konto("Jan Kowalski", "123456", 10000, 0, 0)
konto_vip = Konto_VIP("Pawel Nowak", "098765", 3000, 0, 0, )

konto_3.zaloz_lokata(200)
konto_vip.przelew(1000, konto_3)
konto_vip.zaloz_lokata(200)
print(konto_3)
print(konto_vip)
print()
konto_3.wez_kredyt(500)
print(konto_3)

print(konto_vip.historia_konta())
```

```python
from datetime import date
class Pokoj:
    def __init__(self, numer_pokoju :int, typ_pokoju: str, cena_za_noc: float, czy_dostepny: bool):
        self.numer_pokoju = numer_pokoju
        self.typ_pokoju = typ_pokoju
        self.cena_za_noc = cena_za_noc
        self.czy_dostepny = czy_dostepny

    def zarezerwuj_pokoj(self):
        if self.czy_dostepny:
            self.czy_dostepny = False
            return True
        else:
            return False

    def zwolnij_pokoj(self):
        if not self.czy_dostepny:
            self.czy_dostepny = True
            return True
        else:
            return False

    def __str__(self):
        return f"Nr pokoju: {self.numer_pokoju}, Typ Pokoju: {self.typ_pokoju}, Cena za noc: {self.cena_za_noc}, Czy dostepny: {self.czy_dostepny}"

class Gosc:
    def __init__(self, id_goscia: int, imie_nazwisko: str, email: str, telefon: int):
        self.id_goscia = id_goscia
        self.imie_nazwisko = imie_nazwisko
        self.email = email
        self.telefon = telefon

    def __str__(self):
        return f"ID: {self.id_goscia}, Imię i Nazwisko: {self.imie_nazwisko}, Email: {self.email}, Telefon: {self.telefon}"

class Rezerwacja:
    def __init__(self, id_rezerwacji : int, gosc : Gosc, pokoj : Pokoj, data_zameldowania : date, data_wymeldowania : date, calkowita_cena : float):
        self.id_rezerwacji = id_rezerwacji
        self.gosc = gosc
        self.pokoj = pokoj
        self.data_zameldowania = data_zameldowania
        self.data_wymeldowania = data_wymeldowania
        self.calkowita_cena = calkowita_cena

    def oblicz_calkowita_cene(self):
        self.calkowita_cena = self.pobierz_dlugosc_pobytu() * self.pokoj.cena_za_noc

    def pobierz_dlugosc_pobytu(self):
        dlugosc = self.data_wymeldowania - self.data_zameldowania
        return dlugosc.days

    def __str__(self):
        return f"ID: {self.id_rezerwacji}, Gosc: {self.gosc.id_goscia} {self.gosc.imie_nazwisko}, Pokoj: {self.pokoj.numer_pokoju}, Data Zameldowania: {self.data_zameldowania}, Data Wymeldowania: {self.data_wymeldowania}, Calkowita cena: {self.calkowita_cena}"

class Hotel:
    def __init__(self, nazwa : str, pokoje : list, rezerwacje : list):
        self.nazwa = nazwa
        self.pokoje = pokoje
        self.rezerwacje = rezerwacje

    def dodaj_pokoj(self, pokoj):
        if pokoj not in self.pokoje:
            self.pokoje.append(pokoj)
            print(f"Pokoj nr {pokoj.numer_pokoju} został dodany do listy pokoi")
        else:
            print(f"Pokoj nr {pokoj.numer_pokoju} juz istnieje" )

    def usun_pokoj(self, pokoj):
        if pokoj.czy_dostepny == 1:
            self.pokoje.remove(pokoj)
            print(f"Pokoj numer: {pokoj.numer_pokoju} został usuniety")
        else:
            print(f"Pokoj numer: {pokoj.numer_pokoju} jest zajęty i nie moznna go usunac.")

    def utworz_rezerwacje(self, gosc, numer_pokoju, data_zameldowania, data_wymeldowania):
        pokoj = next((p for p in self.pokoje if p.numer_pokoju == numer_pokoju), None)

        if not pokoj:
            print(f"Pokoj numer: {numer_pokoju} nie istnieje,")
            return None

        if not pokoj.czy_dostepny:
            print(f"Pokoj numer: {numer_pokoju} jest juz zajety.")
            return None

        pokoj.zarezerwuj_pokoj()

        id_rezerwacji = len(self.rezerwacje) + 1
        rezerwacja = Rezerwacja(id_rezerwacji, gosc, pokoj, data_zameldowania, data_wymeldowania, 0)
        rezerwacja.oblicz_calkowita_cene()

        self.rezerwacje.append(rezerwacja)
        print(f"Utworzono rezerwacje {id_rezerwacji} dla goscia {gosc.imie_nazwisko}")
        return rezerwacja

    def anuluj_rezerwacje(self, id_rezerwacji):
        rezerwacja = next((r for r in self.rezerwacje if r.id_rezerwacji == id_rezerwacji), None)

        if not rezerwacja:
            print(f"Rezerwacja nr {id_rezerwacji} nie istnieje.")
            return False

        rezerwacja.pokoj.zwolnij_pokoj()

        self.rezerwacje.remove(rezerwacja)
        print(f"Rezerwacja nr {id_rezerwacji} zostala usunieta.")
        return True

    def pobierz_rezerwacje_goscia(self, id_goscia):
        rezerwacje_goscia = [r for r in self.rezerwacje if r.gosc.id_goscia == id_goscia]

        if not rezerwacje_goscia:
            print(f"Gosc o ID {id_goscia} nie ma żadnych rezerwacji.")
            return None
        else:
            print(f"Rezerwacje goscia o ID {id_goscia}:")
            for r in rezerwacje_goscia:
                print(f"[{r}]")

p1 = Pokoj(101, "Jednoosobowy", 200.0, True)
p2 = Pokoj(102, "Dwuosobowy", 300.0, True)
hotel = Hotel("Hotel", [p1, p2], [])

g1 = Gosc(1, "Jan Kowalski", "jan@kowalski.pl", 123456789)
g2 = Gosc(2, "Anna Nowak", "anna@nowak.pl", 987654321)

hotel.utworz_rezerwacje(g1, 101, date(2025, 10, 14), date(2025, 10, 17))
hotel.utworz_rezerwacje(g2, 102, date(2025, 10, 15), date(2025, 10, 18))

hotel.pobierz_rezerwacje_goscia(1)
hotel.anuluj_rezerwacje(1)
hotel.pobierz_rezerwacje_goscia(1)

hotel.pobierz_rezerwacje_goscia(2)
hotel.dodaj_pokoj(p1)
hotel.usun_pokoj(p1)
```

```python
class Pojazd:
    def __init__(self, marka: str, model: str, rok_produkcji: int, cena: float = 0):
        self.marka = marka
        self.model = model
        self.rok_produkcji = rok_produkcji
        self.cena = cena

    def __str__(self):
        return f"{self.marka} {self.model} {self.rok_produkcji} {self.cena}"

    def oblicz_cena(self):
        self.cena = 10 * self.rok_produkcji
    
class Samochod(Pojazd):
    def __init__(self, marka: str, model: str, rok_produkcji: int, cena: float = 0, ma_klimatyzacje: bool = True):
        super().__init__(marka, model, rok_produkcji, cena)
        self.ma_klimatyzacje = ma_klimatyzacje

    def oblicz_cena(self):
        self.cena = 10 * self.rok_produkcji + (2000 if self.ma_klimatyzacje else 0)

    def __str__(self):
        return f"{super().__str__()} {self.ma_klimatyzacje}"

auto = Samochod("Opel", "Vectra", 2007)
auto.oblicz_cena()

motor = Pojazd("abc", "def", 2007)
motor.oblicz_cena()

print(auto)
print(motor)
```
