```python
def zad6(a):
    f = 2
    wynik = []
    while f*f <= a:
        while a % f == 0:
            wynik.append(f)
            a //= f
        f+=1
    if a > 1:
        wynik.append(a)
    return wynik
print(zad6(120))
```

```python
def nwd(a, b):
    while b > 0:
        temp = a % b
        a = b
        b = temp % b
    return a
def zad7(a, b):
    return a*b/nwd(a, b)

print (zad7(4,6))
```

```python
def zad9(a,b,c):
    delta = b*b - 4*a*c
    if a!= 0:
        if delta < 0:
            return False
        elif delta == 0:
            x1 = (-b / (2*a))
            return x1
        else:
            x1 = (-b + math.sqrt(delta)) / (2*a)
            x2 = (-b - math.sqrt(delta)) / (2*a)
            return x1, x2
    else:
        return False
print(zad9(-2,4,-2))
```

```python
def zad18(text):
    wynik = []
    n = len(text)

    def rekurencja(start, obecne):
        if start == n:
            wynik.append(obecne[:])
            return

        for i in range(start + 1, n + 1):
            fragment = text[start:i]
            rekurencja(i, obecne+ [fragment])

    rekurencja(0, [])
    return wynik

text = "pies"
print(zad18(text))
```

```python
def odwrM(a, m):
    for i in range(1, m):
        if (a * i) % m == 1:
            return i

def zad26(tekst, a, b):
    m = 26
    wynik = ""
    alfabet = "abcdefghijklmnopqrstuvwxyz"
    aOdwr = odwrM(a, m)

    for x in tekst:
        if x in alfabet:
            l = alfabet.index(x)
            odszyfr = (aOdwr * (l - b)) % m
            wynik += alfabet[odszyfr]
        else:
            wynik += x
    return wynik

szyfr = zad25(tekst,7,4)
zad26(szyfr,7,4)
```

```python
def zad27(file_in):
    with open(file_in, "r") as f:
        text = f.read()
        freq = {}
        for word in text:
            freq[word] = freq.get(word, 0) + 1
        probs = [count / len(text) for count in freq.values()]
        entr = -sum(p * math.log2(p) for p in probs)
        return entr
print(zad27("data/in.txt"))
```

```python
def zad28(file_in, file_out):
    with open(file_in, "r") as f:
        text = f.read()
        text = text.split()
        unique = []
        for word in text:
            if word not in unique:
                unique.append(word)
    with open(file_out, "w") as f:
        f.write(" ".join(unique))

zad28("data/in.txt", "data/out.txt")
```

```python
def zad29(text):
    text = " ".join(text.strip().split())
    text = text.split()
    for i in range(len(text)):
        for letter in text[i]:
            if letter == ".":
                text[i+1] = text[i+1].capitalize()
    text = " ".join(text)
    return text

print(zad29("as dajs das.         asd asd asd    asd"))
```

```python
def zad30(file_in):
    with open(file_in, "r") as f:
        text = f.read()
        text = text.split()
        if text[1] == "+":
            return int(text[0]) + int(text[2])
        elif text[1] == "-":
            return int(text[0]) - int(text[2])
        elif text[1] == "*":
            return int(text[0]) * int(text[2])
        elif text[1] == "/":
            if text[2] != "0":
                return int(text[0]) / int(text[2])
            else: return False
        elif text[1] == "**":
            return int(text[0]) ** int(text[2])
print(zad30("data/in30.txt"))
```
