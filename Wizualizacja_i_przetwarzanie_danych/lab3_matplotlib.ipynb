```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def zad1():
    x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
    y1 = np.sin(x)
    plt.plot(x, y1, label='sin(x)')

    extrema_x = np.array([-3*np.pi/2, -np.pi/2, np.pi/2, 3*np.pi/2])
    extrema_y = np.sin(extrema_x)
    plt.scatter(extrema_x, extrema_y, color='red', s=80, label='Ekstrema', zorder=5)

    plt.title('Wykres sin(x) z ekstremami')
    plt.xlabel('x')
    plt.ylabel('y')

    plt.xlim(-2 * np.pi, 2 * np.pi)
    plt.ylim(-1.25, 1.25)

    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    plt.legend()

    plt.show()
zad1()
```

```python
def zad2():
    n = 100
    x = np.random.rand(n)
    y = np.random.rand(n)
    sizes = np.random.randint(200, 2000, size=n)

    plt.figure(figsize=(12, 8))
    scatter = plt.scatter(x, y, c = (x*y), cmap = "hsv", s=sizes, alpha=.8)
    plt.colorbar(scatter)
    plt.show()
zad2()
```

```python
def zad3():
    x = np.linspace(-2*np.pi, 2*np.pi, 500)

    funkcje = [
        (lambda x: np.sin(x), "sin(x)"),
        (lambda x: np.cos(x), "cos(x)"),
        (lambda x: np.tan(x), " tg(x)"),
        (lambda x: 1/np.tan(x), "ctg(x)"),
    ]

    fig, ax = plt.subplots(2, 2, figsize=(12, 8))

    for ax, (f, opis) in zip(ax.flat, funkcje):
        y = f(x)
        y = np.ma.masked_where(np.abs(y) > 10, y)
        ax.plot(x, y)
        ax.set_title(f"f(x) = {opis}")
        ax.grid(True)
        ax.axhline(0, color='black')
        ax.axvline(0, color='black')
        ax.set_xlabel('x')
        ax.set_ylabel(opis)
    plt.show()
zad3()
```

```python
def zad4():
    dane = {
        'matematyka': {
            'kolokwium': [4, 5, 3.5],
            'egzamin': [3.5, 5, 4]
        }
    }
    kolokwium = dane['matematyka']['kolokwium']
    egzamin = dane['matematyka']['egzamin']
    liczba_studentow = len(kolokwium)

    nr_studenta = np.arange(liczba_studentow) + 1
    width = 0.2
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.bar(nr_studenta - width/2 - 0.05, kolokwium, width, label='Kolokwium')
    ax.bar(nr_studenta + width/2 + 0.05, egzamin, width, label='Egzamin')

    ax.set_xlabel("Student")
    ax.set_ylabel("Ocena")
    ax.set_xticks(nr_studenta)
    ax.legend()
    plt.show()
zad4()
```

```python
def zad5():
    n = 1000
    x_rand = np.random.uniform(0, np.pi, n)
    y_rand = np.random.uniform(0, 1, n)

    sin = np.sin(x_rand)

    inside_x = x_rand[y_rand <= sin]
    inside_y = y_rand[y_rand <= sin]

    outside_x = x_rand[y_rand > sin]
    outside_y = y_rand[y_rand > sin]

    x = np.linspace(0, np.pi, 200)
    plt.figure(figsize=(8, 6))
    plt.plot(x, np.sin(x),c = "black")

    plt.scatter(inside_x, inside_y, color='red')
    plt.scatter(outside_x, outside_y, color='blue')

    plt.title('Pole pod funkcj sin(x)')
    plt.xlabel('x')
    plt.ylabel('sin(x)')
    plt.grid()
    plt.xlim(0, np.pi)
    plt.ylim(0, 1)

    plt.show()

zad5()
```
