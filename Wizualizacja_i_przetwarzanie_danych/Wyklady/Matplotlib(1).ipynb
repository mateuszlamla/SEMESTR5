### Matplotlib

[Matplotlib](https://matplotlib.org/stable/index.html) jest biblioteką służącą do tworzenia grafiki 2D o dużej jakości. Może być użyta m.in. w skryptach Pythona czy notebookach Jupytera.

Najczęściej z `matplotlib` wybieramy moduł `pyplot`.

`import matplotlib.pyplot as plt`

### Plot

Podstawowym poleceniem `pyplot` służącym do rysowania wykresów itp. jest `plot()`.

Aby dodatkowo zobaczyć rysunek używamy komendy `show()`.

Jeśli byśmy chcieli zapisać wykres, to należy podać komendę `savefig('nazwa.rozszerzenie')`, gdzie dostępne rozszerzenia to m.in. `png, svg, eps` czy `pdf`.

````python
import matplotlib.pyplot as plt

x = [1, 2, 3]

plt.plot(x, x)
plt.grid()
plt.show()
````

#### Linie

Domyślnie `plot()` tworzy łamaną przechodzącą przez zadane punkty o ustalonym kolorze. Można to zmienić wprowadzając kilka dodatkowych parametrów.

`plt.plot(x, y, linestyle="rodzaj linii", marker="marker", color = "kolor")`

gdzie `rodzaj linii` można wybrać z następującego zestawu

|Symbol|Rodzaj linii|
|:---|:---|
|-|	linia|
|--| podwójna linia|
|-.| linia-punkt|
|:|	linia punktowana|

Domyślnie linia jest pojedyncza.

#### Markery punktów

Markery można wybrać z następującego zestawienia

|Symbol|Marker|Symbol|Marker|
|:---|:---|:---|:---|
|.|	punkt|,|piksel|
|o|	kółka|v|trójkąt w dół|
|^|	trójkąt w góre|<| trójkąt w lewo|
|>|	trójkąt w prawo|s|kwadrat|
|p|pięciokąt|h|sześciokąt|
|\*|gwiazdki|+|plusy|
|x|x|D/d|diament duży/mały|

#### Kolory

Kolor z kolei wybierzemy z zestawu

|Symbol|Kolor|Symbol|Kolor|
|:---|:---|:---|:---|
|b|niebieski|g|zielony|
|r|czerwony|c|cyjan|
|m|	magenta|y|żółty|
|k|	czarny|w|biały|

Można również podać: 
+ pełne nazwy barw, np. `'red'` 
+ w postaci szesnastkowej, np. `'#008000'`
+ krotki RGB or RGBA, np. `(0,1,0,1)`
+ odcienie szarości jako string, np.  `'0.8'`

````python
x = [1, 2, 3]
y = [3, 2 , 1]

plt.plot(x, y, marker="d", color="r", linestyle=":")
plt.plot(x, marker=',', linestyle="--", color="green")
plt.plot(x, x, marker='p', color=(0.1, 0.9, 0.6))
plt.plot(y, marker='o', color='0.5')

plt.show()
````

#### Opis wykresu i osi

Osie wykresów można opisać używając do tego m.in. $\LaTeX$a i komendy `xlabel()` (`ylabel()`). Sam wykres może mieć z kolei tytuł za pomocą `title()`. Za pomocą parametru `fontsize` ustalimy wielkość czcionki.

````python
x = [1, 2, 3]
y = [3, 2 , 1]

plt.plot(x, y, marker="d", color="r", linestyle=":")
plt.plot(x, marker=',', linestyle="--", color="green")
plt.plot(x, x, marker='p', color=(0.1, 0.9, 0.6))
plt.plot(y, marker='o', color='0.5')
plt.xlabel('oś $\sin x^2$', fontsize=10)
plt.ylabel('oś $y$', fontsize=15)
plt.title('tytuł', fontsize=20)
plt.show()
````

#### Legenda

W przypadku wielu wykresów na jednym płótnie przydatna może być legenda, którą tworzymy w ten sposób, że każdy `plot()` ma swoją etykietę `label` jako parametr i dodajemy komendę `legend()` z ewentualną lokalizacją.

````python
plt.plot(x, y, marker="d", color="r", linestyle=":", label="wykres 1")
plt.plot(x, marker=',', linestyle="--", color="green", label="wykres 2")
plt.plot(x, x, marker='p', color=(0.1, 0.9, 0.6), label="wykres 3")
plt.plot(y, marker='o', color='0.5', label="wykres 4")
plt.xlabel('oś $x$', fontsize=10)
plt.ylabel('oś $y$', fontsize=15)
plt.title('tytuł', fontsize=20)
plt.legend(loc="center right")
plt.show()
````

Lokalizację legendy można wyznaczyć za pomocą wartości

|Pozycja| |
|:---|:---|
|best|upper right|
|upper left|lower left|
|lower right|right|
|center left|center right|
|lower center|upper center|
|center| |

### Zakres i wartości osi

Jeśli wykres jest określony na zbyt szerokim obszarze, to można go zawężyć podając komendy `xlim()` i `ylim()`. 

Jeśli natomiast siatka osi nas nie zadowala, to można wprowadzić swoją za pomocą pleceń `xticks()` i `yticks()`.

````python
plt.plot(x, y, marker="d", color="r", linestyle=":", label="wykres 1")
plt.plot(x, marker=',', linestyle="--", color="green", label="wykres 2")
plt.plot(x, x, marker='p', color=(0.1, 0.9, 0.6), label="wykres 3")
plt.plot(y, marker='o', color='0.5', label="wykres 4")
plt.xlabel('oś $x$', fontsize=10)
plt.ylabel('oś $y$', fontsize=15)
plt.title('tytuł', fontsize=20)
plt.legend(loc="best")
plt.xlim(1,2)
plt.ylim(-1,4)
plt.xticks([1, 1.5, 2],["A", "B", "C"])
plt.yticks([1, 1.5, 2])
plt.show()
````

### Kilka wykresów

Jeśli chcemy wygenerować kilka wykresów jednocześnie, to każdy z nich należy potraktować jako oddzielną figurę podając komendę `figure()`. Dodatkowo można określić tak rozmiar wykresu za pomocą `figsize` i podać rozdzielczość za pomocą `dpi`. 

````python
plt.figure(figsize=(4,3))
plt.plot(x, y, marker="d", color="r", linestyle=":", label="wykres 1")
plt.plot(x, marker=',', linestyle="--", color="green", label="wykres 2")

plt.figure(figsize=(3,4), dpi=25)
plt.plot(x, x, marker='p', color=(0.1, 0.9, 0.6), label="wykres 3")
plt.plot(y, marker='o', color='0.5', label="wykres 4")

plt.show()
````

### Funkcje $ \sin (x)$ i $ \cos (x)$ jako `subploty`

##### Wersja 1

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,2*np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x, y1)
plt.xlabel("x", fontsize=15)
plt.ylabel("$\sin x$", fontsize=15)
plt.xlim(0,2*np.pi)
plt.grid()

plt.subplot(2,1,2)
plt.plot(x, y2)
plt.xlabel("x", fontsize=15)
plt.ylabel("$\cos x$", fontsize=15)
plt.xlim(0,2*np.pi)
plt.grid()

plt.tight_layout()
plt.show()
````

##### Wersja 2

````python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

ax1.plot(x, y1)
ax1.set_xlabel("x", fontsize=15)
ax1.set_ylabel("$\sin x$", fontsize=15)
ax1.set_xlim(0,2*np.pi)
ax1.grid()

ax2.plot(x, y2)
ax2.set_xlabel("x", fontsize=15)
ax2.set_ylabel("$\cos x$", fontsize=15)
ax2.set_xlim(0,2*np.pi)
ax2.grid()

plt.tight_layout()

plt.show()
````

##### Wersja 3

````python
fig, axs = plt.subplots(nrows=2, ncols=1)

axs = axs.ravel()

x = np.linspace(0,2*np.pi,100)
y = [np.sin(x), np.cos(x)]

etykiety_y = ["$\sin x$", "$\cos x$"]

for i in range(2):
    axs[i].plot(x, y[i])
    axs[i].set_xlabel("$x$", fontsize=15)
    axs[i].set_ylabel(etykiety_y[i], fontsize=15)
    axs[i].set_xlim(0,2*np.pi)
    axs[i].grid()

plt.tight_layout()

plt.show()
````

### `bar()`

`bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)`

````python
import numpy as np

studenci = (550, 650, 2200)
    
plt.bar(range(len(studenci)), studenci, width = 0.5, color='r')

plt.title('Wirtualne zestawienie studentów')
plt.ylabel('Ilość')
plt.xticks(np.arange(3), ('RMS', 'RE', 'RMT'))
plt.show()
````

````python
import numpy as np

studenci = (200, 600, 1000)
studentki = (350, 50, 1200)
    
plt.bar(range(len(studenci)), studenci, width = 0.5, color='r', label="studenci")
plt.bar(range(len(studentki)), studentki, width=0.5,
              bottom=studenci, label="studentki")

plt.title('Wirtualne zestawienie studentów')
plt.ylabel('Ilość')
plt.xticks(np.arange(3), ('RMS', 'RE', 'RMT'))
plt.yticks(np.arange(0, 2400, 300))
plt.legend(loc="best")
plt.show()
````

### Pie

````python
import matplotlib.pyplot as plt
 
labels = ['A', 'B', 'C', 'D', 'E', 'F']
wyniki = [235,134,49,30,11,1]

plt.pie(wyniki, labels=labels)
 
plt.show()
````

````python
kolory = ['red', 'green', 'blue', 'yellow', 'black', 'pink']
explode = (0.4, 0.2, 0.3, 0.9, 0.1, 0.9)  
 
plt.pie(wyniki, explode=explode, labels=labels, colors=kolory,
         autopct=lambda p: '{:,.0f}\n ({:.2f}%) '.format(p * sum(wyniki)/100, p), shadow=True, startangle=200)
 
plt.axis('equal')
plt.show()
````

### Table
[https://matplotlib.org/api/_as_gen/matplotlib.pyplot.table.html#matplotlib.pyplot.table](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.table.html#matplotlib.pyplot.table)

````python
import numpy as np
dane = [[ 235, 138, 28, 0, 16, 42, 0, 1], [235, 134, 0, 49,  30, 0, 11, 1]]

kolumny = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
wiersze = ['2024', '2025']
kolory = ['red', 'blue']
width = 0.25

plt.figure(figsize=(20,4))

plt.bar(np.arange(len(kolumny))- width/2, dane[0], width=width, color=kolory[0], label=wiersze[0])
plt.bar(np.arange(len(kolumny))+ width/2, dane[1], width=width,
              color=kolory[1], label=wiersze[1])
plt.xticks([])
plt.yticks(np.arange(0,250,50))

tabela = plt.table(cellText=dane,
                      rowLabels=wiersze,
                      rowColours=kolory,
                      colLabels=kolumny,
                      loc='bottom')
tabela.set_fontsize(13)

plt.subplots_adjust(left=0.2, bottom=0.)
plt.title('Zestawienie')

plt.show()
````
