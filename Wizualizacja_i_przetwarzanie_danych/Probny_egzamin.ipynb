```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data/gastronomia.xlsx', sheet_name='TABLICA')
df
```

```python
df.columns = ['Kod', 'Nazwa', 'zeSprzedazyTowarowHandlowych2021', 'zeSprzedazyTowarowHandlowych2022', 'zProdukcjiGastronomicznej2021', 'zProdukcjiGastronomicznej2022', 'zeSprzedazyNapojowAlkiWyrobowTyton2021', 'zeSprzedazyNapojowAlkiWyrobowTyton2022']
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df['Kod'] = df['Kod'].astype(int)
df['zeSprzedazyTowarowHandlowych2021'] = df['zeSprzedazyTowarowHandlowych2021'].astype(float)
df['zeSprzedazyTowarowHandlowych2022'] = df['zeSprzedazyTowarowHandlowych2022'].astype(float)
df['zProdukcjiGastronomicznej2021'] = df['zProdukcjiGastronomicznej2021'].astype(float)
df['zProdukcjiGastronomicznej2022'] = df['zProdukcjiGastronomicznej2022'].astype(float)
df['zeSprzedazyNapojowAlkiWyrobowTyton2021'] = df['zeSprzedazyNapojowAlkiWyrobowTyton2021'].astype(float)
df['zeSprzedazyNapojowAlkiWyrobowTyton2022'] = df['zeSprzedazyNapojowAlkiWyrobowTyton2022'].astype(float)
dfNoSum = df.copy()
```

```python
sums = df.select_dtypes(include=np.number).sum()
sums['Kod'] = 0
sums['Nazwa'] = 'SUMA'
df.loc[len(df)] = sums
df
```

```python
slask2021_1 = df[df['Nazwa'] == 'ŚLĄSKIE']['zProdukcjiGastronomicznej2021'].sum()
slask2022_1 = df[df['Nazwa'] == 'ŚLĄSKIE']['zProdukcjiGastronomicznej2022'].sum()

slask2021_2 = df[df['Nazwa'] == 'ŚLĄSKIE']['zeSprzedazyTowarowHandlowych2021'].sum()
slask2022_2 = df[df['Nazwa'] == 'ŚLĄSKIE']['zeSprzedazyTowarowHandlowych2022'].sum()

slask2021_3 = df[df['Nazwa'] == 'ŚLĄSKIE']['zeSprzedazyNapojowAlkiWyrobowTyton2021'].sum()
slask2022_3 = df[df['Nazwa'] == 'ŚLĄSKIE']['zeSprzedazyNapojowAlkiWyrobowTyton2022'].sum()

fig, ax = plt.subplots(1,3, figsize=(18,6))
ax[0].pie([slask2021_1, slask2022_1], labels = ['2021', '2022'], autopct='%1.1f%%', startangle=90)
ax[0].set_title('Udział produkcji gastronomicznej')

ax[1].bar(['Handlowe', 'Alkoholowe'], [slask2021_2, slask2021_3], color=['blue', 'orange'])
ax[1].set_title('2021')

ax[2].bar(['Handlowe', 'Alkoholowe'], [slask2022_2, slask2022_3], color=['blue', 'orange'])
ax[2].set_title('2022')

plt.show()


```

```python
maxAlk2022 = dfNoSum[['Nazwa', 'zeSprzedazyNapojowAlkiWyrobowTyton2022']].loc[dfNoSum['zeSprzedazyNapojowAlkiWyrobowTyton2022'].idxmax()]
maxAlk2021 = dfNoSum[['Nazwa', 'zeSprzedazyNapojowAlkiWyrobowTyton2021']].loc[dfNoSum['zeSprzedazyNapojowAlkiWyrobowTyton2021'].idxmax()]
maxHandl2022 = dfNoSum[['Nazwa', 'zeSprzedazyTowarowHandlowych2022']].loc[dfNoSum['zeSprzedazyTowarowHandlowych2022'].idxmax()]
maxHandl2021 = dfNoSum[['Nazwa', 'zeSprzedazyTowarowHandlowych2021']].loc[dfNoSum['zeSprzedazyTowarowHandlowych2021'].idxmax()]
maxGastr2022 = dfNoSum[['Nazwa', 'zProdukcjiGastronomicznej2022']].loc[dfNoSum['zProdukcjiGastronomicznej2022'].idxmax()]
maxGastr2021 = dfNoSum[['Nazwa', 'zProdukcjiGastronomicznej2021']].loc[dfNoSum['zProdukcjiGastronomicznej2021'].idxmax()]
print("Maksymalna sprzedaż napojów alkoholowych w 2022 roku:")
print(maxAlk2022)
print("\nMaksymalna sprzedaż napojów alkoholowych w 2021 roku:")
print(maxAlk2021)
print("\nMaksymalna sprzedaż towarów handlowych w 2022 roku:")
print(maxHandl2022)
print("\nMaksymalna sprzedaż towarów handlowych w 2021 roku:")
print(maxHandl2021)
print("\nMaksymalna produkcja gastronomiczna w 2022 roku:")
print(maxGastr2022)
print("\nMaksymalna produkcja gastronomiczna w 2021 roku:")
print(maxGastr2021)

```

```python
dfNoSum['Iloraz_zProdukcjiGastronomicznej'] = dfNoSum['zProdukcjiGastronomicznej2022'] / dfNoSum['zProdukcjiGastronomicznej2021']
df_sorted = dfNoSum.sort_values('Iloraz_zProdukcjiGastronomicznej', ascending=False)
plt.figure(figsize=(10,5))
plt.bar(df_sorted['Nazwa'], df_sorted['Iloraz_zProdukcjiGastronomicznej'], color='skyblue')
plt.xlabel('Województwo')
plt.ylabel('Iloraz produkcji gastronomicznej (2022/2021)')
plt.title('Iloraz produkcji gastronomicznej w 2022 roku do 2021 roku')
plt.xticks(rotation=90)
plt.show()
```
