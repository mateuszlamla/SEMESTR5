```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
with open('data/piłkarz.json') as f:
    data = json.load(f)

df = pd.DataFrame(data)
df
```

ZADANIE 1

```python
df['podsumowanie'] = [f"{s['mecze']}-{s['gole']}" for s in df['suma']]
df['podsumowanie']
```

ZADANIE 2

```python
df_primera = df[df['liga'] == 'Primera División']

df_primera2223 = df_primera[df_primera['sezon'] == '2022/2023']
df_primera2324 = df_primera[df_primera['sezon'] == '2023/2024']
df_primera2425 = df_primera[df_primera['sezon'] == '2024/2025']

fig, ax = plt.subplots(3, 1, figsize=(10,10))

mecze = df_primera2223['mecze'].iloc[0]
gole = df_primera2223['gole'].iloc[0]
ax[0].bar(['Mecze', 'Gole'], [df_primera2223['mecze'].iloc[0], df_primera2223['gole'].iloc[0]], color=['blue', 'magenta'])
ax[0].set_title('Sezon 2022/2023')
ax[0].set_ylabel('Ilość')

ax[1].bar(['Mecze', 'Gole'], [df_primera2324['mecze'].iloc[0], df_primera2324['gole'].iloc[0]], color=['blue', 'magenta'])
ax[1].set_title('Sezon 2023/2024')
ax[1].set_ylabel('Ilość')

ax[2].bar(['Mecze', 'Gole'], [df_primera2425['mecze'].iloc[0], df_primera2425['gole'].iloc[0]], color=['blue', 'magenta'])
ax[2].set_title('Sezon 2024/2025')
ax[2].set_ylabel('Ilość')

plt.show()
```

ZADANIE 3

```python
df_ratio = df[df['gole']/df['mecze'] > 0.5]
df_ratio[['sezon', 'liga']]
```

ZADANIE 4

```python
mean_goals_career = df['gole'].mean()
print(mean_goals_career)
df_above_mean = df[df['gole'] > mean_goals_career]
df_above_mean
```

ZADANIE 5

```python
clubs = df['klub'].unique()
seasons_in_club = {}
for club in clubs:
    seasons_in_club[club] = df[df['klub'] == club]['sezon'].nunique()

plt.figure(figsize=(10,10))
plt.bar(seasons_in_club.keys(), seasons_in_club.values())
plt.xticks(rotation=45)
plt.xlabel('Klub')
plt.ylabel('Liczba sezonów')
plt.title('Liczba sezonów spędzonych w każdym klubie')
plt.show()
```
