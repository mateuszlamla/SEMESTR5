```python
df = pd.read_html("https://www.bankier.pl/inwestowanie/profile/quote.html?symbol=WIG20")
df = df[1]

df['Zmiana procentowa'] = (
    df['Zmiana procentowa'].astype(str).str.replace("%","").str.replace(",",".").astype(float)
)

df['Kurs'] = (
    df['Kurs'].astype(str).str.replace(" ", "").str.replace(",", ".").astype(float)
)

max_zmiana_procent = df.loc[abs(df["Zmiana procentowa"]).idxmax()]
min_zmiana_procent = df.loc[abs(df["Zmiana procentowa"]).idxmin()]
print(f"Najwieksza zmiana procentowa: {max_zmiana_procent['Nazwa']}")
print()
print(f"Najmniejsza zmiana procentowa: {min_zmiana_procent['Nazwa']}")

wybrane = df[df['Nazwa'].isin(['ALIOR', 'PEKAO', 'PKOBP'])]

plt.figure(figsize=(10,10))
plt.bar(wybrane['Nazwa'], wybrane['Kurs'], color= ['orange', 'red', 'blue'])
plt.title("Kursy wybranych spolek")
plt.xlabel("Nazwa")
plt.ylabel("Kurs")

plt.show()

df['Znak'] = df['Zmiana'].apply(lambda x: '+' if x > 0 else '-')
print(df[['Nazwa','Znak']])

df = df.sort_values(by='Kurs', ascending=True)
df

```

```python
def zad3():
    df_zloto = pd.read_csv("zloto.csv")
    srednie = df_zloto.mean(numeric_only=True)
    print(srednie)
    print()
    print()

    df_zloto["Data"] = pd.to_datetime(df_zloto["Data"])
    miesiace_8_9 = df_zloto['Data'].dt.month.isin([8,9])
    df_sierpien_wrzesien = df_zloto[miesiace_8_9]
    srednia_otw_zamkn = df_sierpien_wrzesien[['Otwarcie', 'Zamkniecie']].mean()

    plt.figure(figsize=(10,10))
    plt.bar(['Otwarcie', 'Zamkniecie'], srednia_otw_zamkn)
    plt.title("Srednia Otwarcie i Zamkniecie w miesiacach sierpien wrzesien")
    plt.ylabel("Cena")
    plt.show()

    df_zloto['Najwyzszy - Najnizszy'] = df_zloto['Najwyzszy'] - df_zloto['Najnizszy']
    df_sorted = df_zloto.sort_values(by=['Najwyzszy - Najnizszy'], ascending=True)
    print(df_sorted['Najwyzszy - Najnizszy'])
    print()
    print()

    warunek = (df_zloto['Najwyzszy'] - df_zloto['Otwarcie']) > df_zloto['Najwyzszy - Najnizszy']
    ile = df_zloto[warunek].shape[0]
    print(ile)

zad3()
```
