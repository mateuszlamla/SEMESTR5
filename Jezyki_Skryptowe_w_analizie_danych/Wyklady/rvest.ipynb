## Wczytanie strony

````r
library(rvest)

strona = read_html("https://ms.polsl.pl/kmmti/")
````

### Dostęp do wezłów

**Wszystkie linki ze strony**

````r
linki = strona %>%
  html_nodes("a") 

linki
````

**Zawartość linków**

````r
linki_text = linki %>%
  html_text() 

linki_text
````

**Adresy linków**

````r
linki_adres = linki %>%
  html_attr("href") 

linki_adres
````

### Dostęp do węzła za pomocą `id`

````r
tlo = strona %>% 
  html_nodes("#tlo") 

tlo
````

### Dostęp do węzła po klasie

**Pobieramu linki z menu**

````r
linki_menu = strona %>% 
  html_nodes(".menu") %>%
    html_nodes("li") %>%
    html_nodes("a")

linki_menu
````

### Formularze

Biblioteka ta pozwala również na pracę z formularzami za pomocą komend `html_session`, `html_form`, , `set_values` czy `submit_form`. Więcej na tena temat można znaleźć m.in. w tutorialu [https://rvest.tidyverse.org/reference/index.html](https://rvest.tidyverse.org/reference/index.html).
