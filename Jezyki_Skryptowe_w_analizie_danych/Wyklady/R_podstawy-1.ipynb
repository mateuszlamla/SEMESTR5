# Funkcje

### Definicja i wywołanie
 
Funkcja w R to zbiór instrukcji, który ma być realizowany każdorazowo po jego wywołaniu.
Aby zdefiniować funkcję konieczne jest:
+ podanie nazwy funkcji,
+ określenie elementów wejściowych,
+ użycie słowa kluczowego `function`,
+ umieszczenie kodu w ciele funkcji,
+ `return(wyrażenie)` w razie potrzeby w ciele

```
nazwaFunkcji = function(){ciało funkcji}
```
+ Wewnątrz `()` mogą znajdować się argumenty, które domyślnie mają swój porządek

```
nazwaFunkcji = function(argument1, argument2){ciało funkcji}
```
Funkcję wywołujemy poprzez użycie jej nazwy z odpowiednim zestawem argumentów.

**Uwaga**

Jeśli nie ma w ciele funkcji `return` to zwracane jest ostatnie wyrażenie.

````r
kwadrat = function(x){ x^2 }

kwadrat(2)
````

````r
kwadrat_2 = function(x){ return(x^2) }

kwadrat_2(2)
````

````r
kwadrat_3 = function(x) x^2
kwadrat_3(2)
````

**Metoda `is.function()` i składowe funkcji**

````r
kwadrat = function(x) x^2

mode(kwadrat)
is.function(kwadrat)
is.vector(kwadrat)
````

````r
kwadrat = function(x) x^2

formals(kwadrat) # lista parametrów
body(kwadrat) # wyrażenie (ciało funkcji)
print(kwadrat) # cały kod funkcji
body(body)
````

**Funkcje na obiektach**

````r
kwadrat = function(x) x^2

y = 32
kwadrat(y)

kwadrat(1:5)
````

### Wskazówki

+ Funkcje pozwalają programować na dużym poziomie abstrakcji. Tworzymy obliczenia dla jakichś danych wejściowych, a następnie wywołujemy je na konkretnych obiektach.
+ Dobrze napisana funkcja pozwala zaoszczędzić nasz czas - wystarczy jej używać, kiedy chcemy przeprowadzić obliczenia danego typu. Nie musimy powtarzać tych samych fragmentów kodu tam, gdzie chcemy przeprowadzać podobne obliczenia.

### Bloki wyrażeń

+ Rzadko kiedy ciała pisanych funkcji będą składały się z jednego nieskomplikowanego wyrażenia.
+ Do grupowania ciągów wyrażeń (dowolnej ich liczby), które będą traktowane przez język R jako pojedyncze wyrażenia służą - bloki. Tworzymy je poprzez nawiasy klamrowe, tj. `{}`.

````r
wypisz_wektor = function(wekt){
    cat('Wektor ', length(wekt), '-elementowy. Wartości: ', sep='') # funkcja cat to wypisanie wartości ze znakami specjalnymi
    cat(wekt, sep=', ') # sep oznacza separator
    cat('\n') # \n oznacza nową linię
}

wypisz_wektor(1:5)
````

### Zasięg nazw w funkcjach

Przyjrzyjmy się w jaki sposób prowadzone są obliczenia wewnątrz funkcji na obiektach nazwanych, a dokładniej - jakie są relacje międy nazwami tworzonymi wewnątrz funkcji i pozostałymi.

**zmienne lokalne**
+ zarówno argumenty, jak i obiekty nazwane, tworzone wewnątrz funkcji mają zasięg lokalny, czyli nie są widoczne po zakończeniu ewaluacji wywołania funkcji

````r
x = 1

f1 = function(x){
    x = 2
    print(x)
}

f1(x)

x
````

**zmienne globalne**

````r
x = 5

f2 = function(){
    x <<- 7
}

print(x)

print(f2())
print(x)
````

### Parametry i argumenty

**przekazywanie argumentów przez wartość**

````r
f = function(x){
    x[1]=-x[1]
    x
}
y=1:5
f(y)
print(y)
````

**parametry z argumentami domyślnymi**

````r
f1 = function(x=8, y=9){
    cat('x=', x, ' y=', y, '\n', sep='')
}
f1(1,2)        # pełne wywołanie
f1(1)          # pominięty y, tak samo dla f1(x=1)
f1(y=2)        # pominięty x
f1()           # dwa argumenty pominięte
````
