**Dwie serie danych na jednym wykresie - `lines`**

legenda - `legend`

[http://www.sthda.com/english/wiki/add-legends-to-plots-in-r-software-the-easiest-way](http://www.sthda.com/english/wiki/add-legends-to-plots-in-r-software-the-easiest-way)

### Rodzaje wykresów

**Wykres słupkowy `barplot`**

[https://www.rdocumentation.org/packages/graphics/versions/3.6.2/topics/barplot](https://www.rdocumentation.org/packages/graphics/versions/3.6.2/topics/barplot)

````r
dane = c(400, 300, 250, 300) 
barplot(dane)
````

**Wykres słupkowy z dodatkowymi parametrami**

````r
barplot(dane,
main = "Tytuł wykresu", 
xlab = "Oś x", 
ylab = "Os y", 
names.arg = c("Etykieta 1", "Etykieta 2", "Etykieta 3", "Etykieta 4"), 
col = "darkred", 
horiz = TRUE) #układ poziomy
````

**Wykres słupkowy - kategorie danych**

````r
dane <- c(17,18,18,17,18,19,18,16,18,18) 
table(dane) #pozwala na zliczenie ilości wystąpień danej wartości
````

````r
barplot(table(dane),
main="Rozkład danych",
xlab="Oś x",
ylab="Oś y",
border="red", 
col="blue", 
)
````

**wykres słupkowy - macierze**

````r
dane = matrix(rbind(c(142, 137, 201, 125),c(243, 218, 171, 199)), 
              nrow = 2, ncol = 4, dimnames = list(c("Nie", "Tak"), 
              c("a", "b", "c", "d")))
dane
````

````r
barplot(dane,
main = "Zestawienie",
xlab = "Klasy",
col = c("red","green")
)
legend("topright",
c("Nie", "Tak"),
fill = c("red","green")
)
````

````r
barplot(dane,
main = "Zestawienie",
xlab = "Klasy",
beside = TRUE,        
col = c("red","green")
)
legend("topright",
c("Nie","Tak"),
fill = c("red","green")
)
````

**Histogram `hist`**

[https://www.rdocumentation.org/packages/graphics/versions/3.6.2/topics/hist](https://www.rdocumentation.org/packages/graphics/versions/3.6.2/topics/hist)

````r
dane = rnorm(100) #wartości z rozkładu N(0,1)
hist(dane)
````

````r
hist(dane,
main="Rozkład normalny",
xlab="x",
xlim=c(-4,4),
col="darkmagenta",
freq=FALSE #gęstość
)
````

````r
hist(dane, breaks=3)
hist(dane, breaks=20)
````

````r
hist(dane,breaks=c(-4,-2,2,4))
````

**Wykres kołowy `pie`**

[https://www.rdocumentation.org/packages/graphics/versions/3.6.2/topics/pie](https://www.rdocumentation.org/packages/graphics/versions/3.6.2/topics/pie)

````r
dane = c("a" = 400, "b" = 300, "c" = 250, "d" = 200, "e" = 100) 
pie(dane)
````

````r
pie(dane,
labels=as.character(dane),
main="Zestawienie wartości", 
col=c("red","orange","yellow","blue","green"), # wybór kolorów kawałków wykresu
border="brown", # kolor obramowania
clockwise=TRUE # układ zgodny ze wskazówkami zegara
)
````

### Przedstawianie danych na większej ilości wykresów

### `ggplot`

**Pierwszy wykres z biblioteką ggplot2**

Komenda `ggplot` jako dane przyjmuje ramki danych. W R możemy je stworzyć za pomocą polecenia `data.frame(jakieś_dane)`. Przykładowo ramka danych zawierająca dane dotyczące wartości funkcji sinus na przedziale $[0,2\pi]$:

````r
x = seq(0,2*pi,0.02)
y = sin(x)
z = cos(x)

dane = data.frame(list("x" = x, "y" = y, "z" = z))
````

Dostęp do danych uzyskujemy m.in. za pomocą operatora $.

````r
dane$x
````

Najprostszy wykres stworzony za pomocą komendy: `ggplot(data = <DATA>, mapping = aes(<MAPPINGS>)) +  <GEOM_FUNCTION>()`, gdzie

+ `mapping = aes(<MAPPINGS>)` oznacza mapowanie odpowiednich zmiennych na wykres
+ `<GEOM_FUNCTION>()` określa jak zmienne mają być wyświetlone, zobacz m.in. [https://ggplot2.tidyverse.org/reference/](https://ggplot2.tidyverse.org/reference/)

Przykładowo można napisać:

````r
ggplot(data = dane, mapping = aes(x, y)) + geom_point()
````

Zadziała również następujący wariant:

````r
plot = ggplot(dane, aes(x,y)) + geom_point()
plot
````

**Dodanie koloru**

````r
plot = ggplot(dane, aes(x,y)) 
plot + geom_point(color="red")
````

````r
plot = ggplot(dane, aes(x,y, color=z)) #kolor jest uzależniony od zmiennej z
plot + geom_point()
````

**Rozmiar punktów**

````r
plot = ggplot(dane, aes(x,y, size=2)) 
plot + geom_point()
````

````r
plot = ggplot(dane, aes(x,y, size=z)) #rozmiar uzależniony od zmiennej
plot + geom_point()
````

**Markery punktów**

````r
dane_shape = data.frame(shape = 0:24)
ggplot(dane_shape, aes(0, 0, shape = shape)) +
  geom_point(aes(shape = shape), size = 5, fill = 'red') +
  scale_shape_identity() +
  facet_wrap(~shape) +
  theme_void()
````

````r
plot = ggplot(dane, aes(x,y)) 
plot + geom_point(shape=14)
````

**Opisy osi i tytuł wykresu**

````r
plot = ggplot(dane, aes(x,y)) 
plot + geom_point(color="red") + labs(title="Wykres", x="Oś x", y="Oś y")
````

````r
plot = ggplot(dane, aes(x,y)) 
plot + geom_point(color="red") + labs(title="Wykres", x="Oś x", y="Oś y") +  theme(plot.title=element_text(size=30, face="bold"), 
                  axis.text.x=element_text(size=15), 
                  axis.text.y=element_text(size=15),
                  axis.title.x=element_text(size=25),
                  axis.title.y=element_text(size=25))
````

**Zapisanie wykresu**

````r
plot = ggplot(dane, aes(x,y)) + 
geom_point(color="red") + labs(title="Sinus", x="Oś x", y="Oś y")

ggsave("sinus.png", height = 6, width = 12)
````

### `qplot` i `quickplot` czyli tradycyjne podejście

[https://www.rdocumentation.org/packages/ggplot2/versions/3.3.0/topics/qplot](https://www.rdocumentation.org/packages/ggplot2/versions/3.3.0/topics/qplot)
```
qplot(
  x,
  y,
  ...,
  data,
  facets = NULL,
  margins = FALSE,
  geom = "auto",
  xlim = c(NA, NA),
  ylim = c(NA, NA),
  log = "",
  main = NULL,
  xlab = NULL,
  ylab = NULL,
  asp = NA,
  stat = NULL,
  position = NULL
)
```

````r
x = seq(0,2*pi, 0.02)
y = cos(x)
````

````r
qplot(
  x,
  y,
  geom = "auto",
  main = "Cosinus",
  xlab = "x",
  ylab = "y"
)
````

**Instalacja**
1. Poprzez `RStudio`: 
 + odpalamy RStudio
 + z Tools wybieramy Install Packages
 + klikamy install
1. Poprzez `conda`:
 + odpalamy `Anaconda Prompt`
 + wyświetlamy listę środowisk poprzez `conda info --envs`
 + wybieramy środowisko z językiem R poprzez `conda activate nazwa_środowiska`
 + wpisujemy `conda install -c conda-forge r-esquisse`

**Jak odpalić Esquisse?**

W `RStudio` wpisujemy:
+ `esquisse::esquisser()` i pojawia nam się okienko do wyboru danych itp. 

Lub

+ 
```
if (interactive()) {
    options("shiny.port" = 8080)
    esquisse::esquisser(viewer = "browser")    
}
``` 
 i pojawia nam się okienko oraz odpala się przeglądarka.
